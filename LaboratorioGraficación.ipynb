```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

datos = pd.read_csv("/content/casasboston.csv")

#print(datos)

df = datos[["RM", "CRIM", "MEDV", "TOWN", "CHAS"]]

df = datos.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS":"RIO CHARLES",
    "RM":"N_HABITACIONES_MEDIO",
    "MEDV":"VALOR_MEDIANO",
    "LSTAT":"PCT_CLASE_BAJA"

  })
print(datos)
df.N_HABITACIONES_MEDIO.plot.hist(color='red')
plt.xlabel('Número de Habitaciones Medio')
plt.ylabel('Frequency')
plt.title('Distribución del Número de Habitaciones Medio')
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(df['N_HABITACIONES_MEDIO'], df['VALOR_MEDIANO'], marker='o', alpha=0.5)
plt.xlabel('Número de Habitaciones Medio')
plt.ylabel('Valor Mediano de las Viviendas')
plt.title('Relación entre el Número de Habitaciones Medio y el Valor Mediano de las Viviendas')

plt.show()
table = pd.crosstab(df['INDICE_CRIMEN'],df['VALOR_MEDIANO'])
print(table)
```
